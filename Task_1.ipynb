# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

## Loading dataset

```python
data = pd.read_csv('API_SP.POP.TOTL.csv')
data
```

## Separate categorical and continuous variables

```python
categorical = data.select_dtypes(include=['object', 'category'])
continuous = data.select_dtypes(include=['number'])
```

## Plotting categorical variables

```python
# Assuming you want to plot the first 10 rows of the categorical column
subset = categorical.head(10)  # Adjust the number of rows as needed

categorical_col = subset.columns[0]  # Replace with the actual column name if needed
plt.figure(figsize=(10, 6))
sns.countplot(data=subset, x=categorical_col)
plt.xticks(rotation=90)
plt.title('Count Plot of Categorical Column (Subset)')
plt.show()
```
